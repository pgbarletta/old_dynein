```julia
function get_tctex(i)
     return svd(hcat(modes_1xdx[res_1xdx_acb_3, i], modes_1ygt[res_1ygt_acb_3, i],
        modes_2ptt[res_2ptt_acb_3, i],  modes_3ftt[res_3ftt_acb_3, i],
        modes_5jpw[res_5jpw_acb_3, i], modes_5wi4[res_5wi4_acb_3, i]))
end
```

```julia
nmodes = 100
svds_tctex = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex = Array{Float64, 1}(undef, nmodes)

for i = 1:nmodes
    svds_tctex[i] = get_tctex(i)
    us_tctex[i] = svds_tctex[i].S[1] / sum(svds_tctex[i].S)
end
```
