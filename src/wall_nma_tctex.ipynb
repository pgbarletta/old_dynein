```julia
function get_tctex_lf(i)
     return svd(hcat(modes_1xdx[res_1xdx_acb_3, i], modes_1ygt[res_1ygt_acb_3, i],
        modes_2plf[res_2plf_acb_3, i],  modes_3flf[res_3flf_acb_3, i],
        modes_5wi4[res_5wi4_acb_3, i]))
end

function get_tctex_23_lf(i)
     return svd(hcat(modes_2plf[res_2plf_acb_3, i],  modes_3flf[res_3flf_acb_3, i]))
end

function get_tctex_23_lb(i)
     return svd(hcat(modes_2plb[res_2plb_acb_3, i],  modes_3flb[res_3flb_acb_3, i]))
end
```

```julia
nmodes = 100
svds_tctex_lf = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_lf = Array{Float64, 1}(undef, nmodes)

svds_tctex_23_lf = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_lf = Array{Float64, 1}(undef, nmodes)

svds_tctex_23_lb = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_lb = Array{Float64, 1}(undef, nmodes)


for i = 1:nmodes
    svds_tctex_lf[i] = get_tctex_lf(i)
    us_tctex_lf[i] = svds_tctex_lf[i].S[1] / sum(svds_tctex_lf[i].S)
    
    svds_tctex_23_lf[i] = get_tctex_23_lf(i)
    us_tctex_23_lf[i] = svds_tctex_23_lf[i].S[1] / sum(svds_tctex_23_lf[i].S)
    
    svds_tctex_23_lb[i] = get_tctex_23_lb(i)
    us_tctex_23_lb[i] = svds_tctex_23_lb[i].S[1] / sum(svds_tctex_23_lb[i].S)
end
```

```julia
plot(us_tctex_lf,
    xticks = [1 ; collect(5:5:20)],
    yticks = 0.:.2:1., ylims = (0, 1.),  
    xlims = (1, 50), 
    label = "TcTex LF",
    grid = true, guidefont = font(16, "Helvetica"), tickfont = font(11, "Helvetica"),
    yaxis = L"u1/U", xaxis = L"Mode \#",
    linecolor = Colors.colorant"#1E90FF", linewidth = 3)

plot!(us_tctex_23_lf,
    label = "23 LF",
    linecolor = Colors.colorant"#00008B", linewidth = 3)

plot!(us_tctex_23_lb,
    label = "23 LB",
    linecolor = Colors.colorant"#DC143C", linewidth = 3)
```

## IC

```julia
# IC 2PG1
res_ic_tctex_c = collect(195:208)
res_ic_tctex_d = collect(209:222)

res_ic_tctex_c_3 = JUMD.triple(res_ic_tctex_c)
res_ic_tctex_d_3 = JUMD.triple(res_ic_tctex_d)
#

svds_tctex_23_c = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_c = Array{Float64, 1}(undef, nmodes)
svds_tctex_23_d = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_d = Array{Float64, 1}(undef, nmodes)

#
dots_3flb_c = Array{Float64, 1}(undef, nmodes)
dots_5wlb_c = Array{Float64, 1}(undef, nmodes)
dots_3flb_d = Array{Float64, 1}(undef, nmodes)
dots_5wlb_d = Array{Float64, 1}(undef, nmodes)

for i = 1:nmodes
    v_2plb_c = modes_2plb[res_ic_tctex_c_3, i] ./ norm(modes_2plb[res_ic_tctex_c_3, i])
    v_3flb_c = modes_3flb[res_ic_tctex_c_3, i] ./ norm(modes_3flb[res_ic_tctex_c_3, i])
    v_5wlb_c = modes_5wlb[res_ic_tctex_c_3, i] ./ norm(modes_5wlb[res_ic_tctex_c_3, i])
    
    v_2plb_d = modes_2plb[res_ic_tctex_d_3, i] ./ norm(modes_2plb[res_ic_tctex_d_3, i])
    v_3flb_d = modes_3flb[res_ic_tctex_d_3, i] ./ norm(modes_3flb[res_ic_tctex_d_3, i])
    v_5wlb_d = modes_5wlb[res_ic_tctex_d_3, i] ./ norm(modes_5wlb[res_ic_tctex_d_3, i])
    
    dots_3flb_c[i] = abs(dot(v_2plb_c, v_3flb_c))
    dots_5wlb_c[i] = abs(dot(v_2plb_c, v_5wlb_c))
    
    dots_3flb_d[i] = abs(dot(v_2plb_c, v_3flb_d))
    dots_5wlb_d[i] = abs(dot(v_2plb_c, v_5wlb_d))
    
    svds_tctex_23_c[i] = svd(hcat(v_2plb_c, v_3flb_c, v_5wlb_c))
    us_tctex_23_c[i] = svds_tctex_23_c[i].S[1] / sum(svds_tctex_23_c[i].S)
    
    svds_tctex_23_d[i] = svd(hcat(v_2plb_d, v_3flb_d, v_5wlb_d))
    us_tctex_23_d[i] = svds_tctex_23_d[i].S[1] / sum(svds_tctex_23_d[i].S)
end
```
