```julia
function get_tctex_lf(i)
     return svd(hcat(modes_1xdx[res_1xdx_acb_3, i], modes_1ygt[res_1ygt_acb_3, i],
        modes_2plf[res_2plf_acb_3, i],  modes_3flf[res_3flf_acb_3, i],
        modes_5wlf[res_5wlf_acb_3, i]))
end

function get_tctex_23_lf(i)
     return svd(hcat(modes_2plf[res_2plf_acb_3, i],  modes_3flf[res_3flf_acb_3, i],
            modes_5wlf[res_5wlf_acb_3, i]))
end

function get_tctex_23_lb(i)
     return svd(hcat(modes_2plb[res_2plb_acb_3, i],  modes_3flb[res_3flb_acb_3, i],
            modes_5wlb[res_5wlb_acb_3, i]))
end
```

```julia
nmodes = 100
svds_tctex_lf = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_lf = Array{Float64, 1}(undef, nmodes)

svds_tctex_23_lf = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_lf = Array{Float64, 1}(undef, nmodes)

svds_tctex_23_lb = Array{SVD{Float64,Float64,Array{Float64,2}}, 1}(undef, nmodes)
us_tctex_23_lb = Array{Float64, 1}(undef, nmodes)


for i = 1:nmodes
    svds_tctex_lf[i] = get_tctex_lf(i)
    us_tctex_lf[i] = svds_tctex_lf[i].S[1] / sum(svds_tctex_lf[i].S)
    
    svds_tctex_23_lf[i] = get_tctex_23_lf(i)
    us_tctex_23_lf[i] = svds_tctex_23_lf[i].S[1] / sum(svds_tctex_23_lf[i].S)
    
    svds_tctex_23_lb[i] = get_tctex_23_lb(i)
    us_tctex_23_lb[i] = svds_tctex_23_lb[i].S[1] / sum(svds_tctex_23_lb[i].S)
end
```

```julia
struct Interaction
    chid_a::String
    chid_b::String
    resi_a::Int64
    resi_b::Int64
    resn_a::String
    resn_b::String
    atm_a::String
    atm_b::String
    scmc_a::String
    scmc_b::String
    interaction::String
    d::Float64
    
    function Interaction(chid_a::AbstractString, chid_b::AbstractString,
        c_resi_a::AbstractString, c_resi_b::AbstractString,
        resn_a::AbstractString, resn_b::AbstractString,
        atm_a::AbstractString, atm_b::AbstractString,
        scmc_a::AbstractString, scmc_b::AbstractString,
        interaction::AbstractString, d::Float64) 
        
        new(chid_a, chid_b, parse(Int64, c_resi_a), parse(Int64, c_resi_b),
            resn_a, resn_b, atm_a, atm_b, scmc_a, scmc_b, interaction, d)
    end

end
```

```julia
function get_ring_chain(fn, chid::Char)
    # Read file
    crudo = readdlm(fn)[8:end, :]
    header = crudo[1, :]
    ancho = 17
    lineas = Array{Any, 2}(undef, 17, 0)
    i = 2
    # Get the Lines corresponding to the desired chain
    while (crudo[i, :][1][1] != chid) 
        i+=1
    end
    while (crudo[i, :][1][1] == chid)
        lineas = hcat(lineas, crudo[i, :])
        i+=1
    end

    # Turn the raw info in an array of Interaction objects
    ninter = size(lineas)[2]
    inter = Array{Interaction, 1}(undef, ninter)
    for i = 1:ninter
        chid_a, c_resi_a, dummy, resn_a = split(lineas[:, i][1], ":")
        chid_b, c_resi_b, dummy, resn_b = split(lineas[:, i][3], ":")
        interaction, sc_mc_ab = split(lineas[:, i][2], ":")
        sc_mc_a, sc_mc_b = split(sc_mc_ab, "_")
        dtce = lineas[:, i][4]

        if(interaction == "PIPISTACK")
            atm_a = "PI"
            atm_b = "PI"
        else
            atm_a = lineas[:, i][7]
            atm_b = lineas[:, i][8]
        end
        inter[i] = Interaction(chid_a, chid_b, c_resi_a, c_resi_b, atm_a, atm_b, 
            resn_a, resn_b, sc_mc_a, sc_mc_b, interaction, dtce)
    end
    
    return inter
end

function sort_contactos!(w::Array{Interaction,1})
    n = length(w)
    v = Array{Int64, 1}(undef, n)
    [ v[i] = w[i].resi_a for i in 1:n ]
    w = w[sortperm(v)]
end
```

```julia
inter_2plb_c = sort_contactos!(get_ring_chain(joinpath(home_tctex, "2plb/ring/ctcto_2plb"), 'C'))
inter_2plb_d = sort_contactos!(get_ring_chain(joinpath(home_tctex, "2plb/ring/ctcto_2plb"), 'D'))

inter_3flb_c = sort_contactos!(get_ring_chain(joinpath(home_tctex, "3flb/ring/ctcto_3flb"), 'C'))
inter_3flb_d = sort_contactos!(get_ring_chain(joinpath(home_tctex, "3flb/ring/ctcto_3flb"), 'D'))

inter_5wlb_c = sort_contactos!(get_ring_chain(joinpath(home_tctex, "5wlb/ring/ctcto_5wlb"), 'C'))
inter_5wlb_d = sort_contactos!(get_ring_chain(joinpath(home_tctex, "5wlb/ring/ctcto_5wlb"), 'D'))

n_c = [ length(inter_2plb_c) ; length(inter_3flb_c) ; length(inter_5wlb_c) ]
n_d = [ length(inter_2plb_d) ; length(inter_3flb_d) ; length(inter_5wlb_d) ]
hcat(n_c, n_d)
```
